# Datasets

```python
# Diret√≥rio base onde est√£o os ficheiros .csv.gz
base_path = r"../documents/mimic-iv-clinical-database-demo-2.2"

# Ficheiros com nomes de vari√°veis
files_to_load = {
    "patients": "hosp/patients.csv.gz",
    "admissions": "hosp/admissions.csv.gz",
    "diagnoses_icd": "hosp/diagnoses_icd.csv.gz",
    "labevents": "hosp/labevents.csv.gz",
    "prescriptions": "hosp/prescriptions.csv.gz",
    "microbiologyevents": "hosp/microbiologyevents.csv.gz",
    "d_icd_procedures": "hosp/d_icd_procedures.csv.gz"
}

# Carrega cada ficheiro e atribui a uma vari√°vel com o mesmo nome
for name, relative_path in files_to_load.items():
    full_path = os.path.join(base_path, relative_path)
    if os.path.exists(full_path):
        print(f"\nüîπ Lendo {name} de: {full_path}")
        df = pd.read_csv(full_path, compression='gzip', low_memory=False)
        globals()[name] = df  # Cria vari√°vel com o nome do ficheiro
        display(df.head())
    else:
        print(f"‚ùå Ficheiro n√£o encontrado: {full_path}")
```
