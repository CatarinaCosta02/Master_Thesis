### Codes of disiases
icd_code = is the International Coding Definitions (ICD) code
icd_version = version 9 (ICD-9) and version 10 (ICD-10)

```python
diagnoses_icd['icd_code'].count()

code_counts = diagnoses_icd.groupby(['icd_version', 'icd_code']).size().reset_index(name='count')
code_counts.columns = ['icd_version', 'icd_code', 'count']

# Mostra o DataFrame com os resultados
display(code_counts)
code_counts.to_csv('code_counts.csv', index=False) 
```

icd_9 4019 =  Unspecified essential hypertension

icd_10 E785 = Hyperlipidemia, unspecified

```python
code_counts = code_counts.sort_values(by=['count'], ascending=True)
display(code_counts)
```

### Excels of codes 10 and 9

```python
path = r"../documents/dados/"

files_to_load = {
    "excluded9": "icu_codes/section111excludedicd9-jan2025_0.xlsx",
    "excluded10": "icu_codes/section111excludedicd10-jan2025_0.xlsx",
    "valid9": "icu_codes/section111validicd9-jan2025_0.xlsx",
    "valid10": "icu_codes/section111validicd10-jan2025_0.xlsx"
}

for name, relative_path in files_to_load.items():
    full_path = os.path.join(path, relative_path)
    if os.path.exists(full_path):
        print(f"\nðŸ”¹ Lendo {name} de: {full_path}")
        
        df = pd.read_excel(full_path)
        
        csv_path = os.path.splitext(full_path)[0] + ".csv"
        df.to_csv(csv_path, index=False)
        
        globals()[name] = df
        display(df.head())
    else:
        print(f"Ficheiro nÃ£o encontrado: {full_path}")
```
